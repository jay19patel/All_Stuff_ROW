```python
red = [1,2,3,4,5,6]
blue = [1,2,3,4,5,6]
sample = ("red1","blue1")
import random
```

```python
data =[]
def who_win():
    ran_num = (random.randint(1,6),random.randint(1,6))
    if ran_num[0] > ran_num[1]:
        data.append("RED")
    elif ran_num[0] == ran_num[1]:
        data.append("TOSS")
    else :
        data.append("BLUE")
```

```python
for _ in range(50):
    who_win()
```

```python
mydicts = {"RED":int(data.count("RED")),"BLUE":int(data.count("BLUE")),"TOSS":int(data.count("TOSS"))}
```

```python
ans = max(mydicts, key=mydicts.get)
```
